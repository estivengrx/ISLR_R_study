# (c) Use the lm() function to perform a multiple linear regression
with mpg as the response and all other variables except name as
the predictors. Use the summary() function to print the results.
Comment on the output. For instance:
i. Is there a relationship between the predictors and the response?

displacement, weight, year and origin are the most valuable variables to make the linear model.

The intercepts of the linear model correspond to the relationships shown in the pairplot in literal a). Positive and negative intercepts correspond to the visual representation.

```powershell
lm.fit <- lm(mpg ~ . - name, data=data) # Excluding name variable from the linear model
summary(lm.fit)
```

iii. What does the coefficient for the year variable suggest?

The coefficient for year is 0.75. This means that for every one unit change in the year predictor variable, given the other variables fixed there is an increment in the mpg response variable of 0.75. Also, there is a positive linear relationship between the predictor and the response variable.
